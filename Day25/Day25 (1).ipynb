```python
import pandas as pd
import numpy as np
data = [
    {"Name": "Asabeneh", "Country":"Finland","City":"Helsinki"},
    {"Name": "David", "Country":"UK","City":"London"},
    {"Name": "John", "Country":"Sweden","City":"Stockholm"}]
df = pd.DataFrame(data)
print(df)
```

```python
weights = [74, 78, 69]
df['Weight'] = weights
df
```

```python
heights = [173, 175, 169]
df['Height'] = heights
print(df)
```

```python
df['Height'] = df['Height'] * 0.01
df
```

```python
# Using functions makes our code clean, but you can calculate the bmi without one
def calculate_bmi ():
    weights = df['Weight']
    heights = df['Height']
    bmi = []
    for w,h in zip(weights, heights):
        b = w/(h*h)
        bmi.append(b)
    return bmi

bmi = calculate_bmi()
```

```python
df['BMI'] = bmi
df
```

```python
df['BMI'] = round(df['BMI'], 1)
print(df)
```
